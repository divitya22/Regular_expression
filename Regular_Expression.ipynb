```python
## Regular expression
```

```python
import re
```

```python
pattern=r"eggs"

if re.match(pattern,"eggseggseggsw"):
  print("match found")
else:
  print("match not found")
```

```python
## re.search
```

```python
pattern=r"eggs"

if re.search(pattern,"eggseggseggsw"):
  print("match found")
else:
  print("match not found")
```

```python
pattern=r"eggs"

if re.findall(pattern,"eggseggseggsw"):
  print("match found")
else:
  print("match not found")
```

```python
## Metacharacter
```

```python
import re
```

```python
# dot metacharcter
pattern=r"gr.y"

if re.match(pattern,"gray"):
  print("Match found")
```

```python
# Carrot metacharacter

pattern=r"^gr.y$"

if re.match(pattern,"grey"):
  print("Match found")
```

```python
# Star metacharacter

pattern=r"eggs(bacon)*"

if re.match(pattern,"eggs"):
  print("match found")
```

```python
# Character class 
# AA1

pattern=r"[A-z][A-Z][0-9]"

if re.search(pattern,"AA1"):
  print("Match found")
```
